```python
from bert_score import score

def calculate_bert_score(reference_text, generated_text):
    P, R, F1 = score([generated_text], [reference_text], lang='en', model_type='bert-base-uncased')
    return {
        'Precision': P.item(),
        'Recall': R.item(),
        'F1 Score': F1.item()
    }

# Example usage
reference_text = """ Hi, Welcome to Chat Doctor! Yes, from what you have described, it appears that you are having Sinusitis. However, you need to be examined to diagnose it finally. So I'd recommend you to visit your GP and get yourself examined. In case the diagnosis is confirmed, you'll require a dose of Antibiotics to ward off the infection. Also, your GP will advise you certain precautions which need to be followed. Hope this information helps. Feel free to ask if you have any doubt. Wishing you a speedy recovery. With warm regards"""
generated_text = """ Most cases of sinusitis, particularly if they are caused by a viral infection, will improve on their own. Your symptoms, such as pain in the head, yellow mucus, and stuffiness, are common signs of sinusitis. If you've noticed that the symptoms seemed to improve but then got worse, or if they persist for more than 10 days without improvement, it might indicate a bacterial infection or a complication which could require treatment from a healthcare professional.

Here are a few scenarios when it would be advisable to see your GP:

1. If the symptoms are severe and not improving after a week.
2. If you experience a high fever, vision changes, swelling around the eyes, or a severe headache.
3. If you have a history of recurrent or chronic sinusitis.
4. If over-the-counter medications are not providing relief.
5. If symptoms return after seeming to improve (sometimes referred to as "double-sickening").

If you're not sure or if you're uncomfortable with your symptoms at any point, it's always best to contact your GP for personalized medical advice. Your GP can provide a proper diagnosis and treatment plan, which may include antibiotics if a bacterial infection is suspected, or other interventions to help alleviate your symptoms and prevent complications."""
scores = calculate_bert_score(reference_text, generated_text)
print(scores)
```
